```python
class VideoConverterClass:

  def __init__(self, pathToVideoFolder = '/content/drive/MyDrive/2025Mag2/ĞĞ˜Ğ /Ğ’Ğ¸Ğ´ĞµĞ¾/', videoImagesFolder = '/content/drive/MyDrive/2025Mag2/ĞĞ˜Ğ /Ğ’Ğ¸Ğ´ĞµĞ¾/imagesFolder/'):
    self.pathToVideoFolder = pathToVideoFolder # Ğ¿ÑƒÑ‚ÑŒ Ğ´Ğ¾ Ğ¿Ğ°Ğ¿ĞºĞ¸ Ñ Ğ²Ğ¸Ğ´ĞµĞ¾
    self.videoImagesFolder = videoImagesFolder # Ğ¿ÑƒÑ‚ÑŒ Ğ´Ğ¾ Ğ¿Ğ°Ğ¿ĞºĞ¸, Ğ´Ğ»Ñ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ Ğ¾Ñ‚Ğ´ĞµĞ»ÑŒĞ½Ğ¾ Ğ¿Ñ€ĞµĞ´ÑĞºĞ°Ğ·Ğ°Ğ½Ğ½Ñ‹Ñ… Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ñ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğ¹

    # ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ĞºĞ°Ğ´Ñ€Ğ¾Ğ² Ğ² ÑĞµĞºÑƒĞ½Ğ´Ñƒ
    self.fps = 24

  def convertToVid(self, folder_name, video_name = 'vidSeg.avi'):
    '''
    ĞšĞ¾Ğ½Ğ²ĞµÑ€Ñ‚Ğ°Ñ†Ğ¸Ñ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğ¹ Ğ² Ğ²Ğ¸Ğ´ĞµĞ¾
    folder_name - ĞĞ°Ğ·Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ¿Ğ°Ğ¿ĞºĞ¸ Ñ ÑĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğ¼Ğ¸ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸ÑĞ¼Ğ¸
    video_name - Ğ½Ğ°Ğ·Ğ²Ğ°Ğ½Ğ¸Ğµ ÑĞ¾Ğ±Ñ€Ğ°Ğ½Ğ½Ğ¾Ğ³Ğ¾ Ğ²Ğ¸Ğ´ĞµĞ¾
    '''
    pathToVideoImages = self.videoImagesFolder + folder_name
    if not os.path.exists(pathToVideoImages):
      print("Try run Mask Video first")
    else:

      images = [img for img in os.listdir(pathToVideoImages) if img.endswith((".jpg", ".jpeg", ".png"))]
      print("Images:", images)

      # Set frame from the first image
      frame = cv2.imread(os.path.join(pathToVideoImages, images[0]))
      height, width, layers = frame.shape

      # Video writer to create .avi file
      pathToVideo = self.pathToVideoFolder + video_name
      video = cv2.VideoWriter(pathToVideo, cv2.VideoWriter_fourcc(*'DIVX'), self.fps, (width, height))

      # Appending images to video
      for image in images:
          video.write(cv2.imread(os.path.join(pathToVideoImages, image)))

      # Release the video file
      video.release()
      cv2.destroyAllWindows()
      print("Video generated successfully!")

  def predictSingleImg(self, img, model, pathToVideoImages, curIm):
    '''
    Predict single img with the model
    img - Ğ˜Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğµ Ğ´Ğ»Ñ ÑĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ°Ñ†Ğ¸Ğ¸
    model - Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ½Ğ°Ñ Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒ Ğ´Ğ»Ñ ÑĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ°Ñ†Ğ¸Ğ¸
    '''
    results = model(img)  # predict on an image

    # Access the results
    for result in results:
      print(type(result.masks))
      if result.masks:
        xy = result.masks.xy  # mask in polygon format
        xyn = result.masks.xyn  # normalized
        masks = result.masks.data  # mask in matrix format (num_objects x H x W)
      # result.show()
      name = pathToVideoImages + "/" + str(curIm).zfill(32) + '.jpg'
      result.save(filename=name)

    return result

  def maskVideoToImgs(self, fileName, folder_name, model):
    '''
    Ğ¡ĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ¸Ñ€Ğ¾Ğ²Ğ°Ñ‚ÑŒ Ğ²Ğ¸Ğ´ĞµĞ¾ Ñ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸ĞµĞ¼ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğ¹ Ğ² Ğ¾Ñ‚Ğ´ĞµĞ»ÑŒĞ½ÑƒÑ Ğ¿Ğ°Ğ¿ĞºÑƒ
    fileName - Ğ¿ÑƒÑ‚ÑŒ Ğ´Ğ¾ Ğ²Ğ¸Ğ´ĞµĞ¾ Ğ´Ğ»Ñ ÑĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ°Ñ†Ğ¸Ğ¸
    folder_name - ĞĞ°Ğ·Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ¿Ğ°Ğ¿ĞºĞ¸ Ğ”Ğ»Ñ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğ¹
    model - Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ½Ğ°Ñ Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒ Ğ´Ğ»Ñ ÑĞµĞ³Ğ¼ĞµĞ½Ñ‚Ğ°Ñ†Ğ¸Ğ¸
    '''

    pathToVideoImages = self.videoImagesFolder + folder_name
    if not os.path.exists(pathToVideoImages):
      os.makedirs(pathToVideoImages)
    else:
      fileList = glob(pathToVideoImages+'/*')
      print("Number of files: ",len(fileList))
      for filePath in fileList:
        try:
            os.remove(filePath)
        except:
            print("Error while deleting file : ", filePath)

    print("Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚Ğ°Ğ½Ğ½Ñ‹Ğµ ĞºĞ°Ğ´Ñ€Ñ‹ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ÑÑ‚ÑÑ Ğ² Ğ¿Ğ°Ğ¿ĞºÑƒ", pathToVideoImages, ". Ğ•ÑĞ»Ğ¸ Ğ¿Ñ€Ğ¾Ğ¸Ğ·Ğ¾ÑˆĞ»Ğ° Ğ¾ÑˆĞ¸Ğ±ĞºĞ°, Ğ·Ğ°Ğ¿ÑƒÑÑ‚Ğ¸Ñ‚Ğµ makeVideo, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ ÑĞ¾Ğ±Ñ€Ğ°Ñ‚ÑŒ ÑĞ¾Ğ±Ñ€Ğ°Ğ½Ğ½Ñ‹Ğµ ĞºĞ°Ğ´Ñ€Ñ‹ Ğ² Ğ²Ğ¸Ğ´ĞµĞ¾")

    cap = cv2.VideoCapture(fileName)
    # Ğ¿Ğ¾Ğ»ÑƒÑ‡ĞµĞ½Ğ¸Ğµ ĞºĞ°Ğ´Ñ€Ğ¾Ğ² Ğ² ÑĞµĞºÑƒĞ½Ğ´Ñƒ
    fps = cap.get(cv2.CAP_PROP_FPS)
    self.fps = fps
    curIm = 0;
    while True:
      success, img = cap.read()
      if not success:
        cap.release()
        cv2.destroyAllWindows()
        break

      imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

      # Ğ¿Ğ¾Ğ»ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ¿Ñ€ĞµĞ´ÑĞºĞ°Ğ·Ğ°Ğ½Ğ¸Ñ
      pred_mask = self.predictSingleImg(imgRGB, model, pathToVideoImages, curIm)
      print(type(pred_mask))

      # name = pathToVideoImages+"/"+str(curIm).zfill(32) + '.jpg'
      # cv2.imwrite(f"{name}", pred_mask)
      curIm += 1
      if curIm % 24 == 0:
        print(curIm)
      cv2.waitKey(1)
    cap.release()
    cv2.destroyAllWindows()
```
